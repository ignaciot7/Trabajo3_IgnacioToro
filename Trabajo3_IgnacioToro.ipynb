```python
from bs4 import BeautifulSoup
```

```python
#Abriendo archivo index.html
with open(r'index.html', "r") as doc_html:
    page = BeautifulSoup(doc_html.read(), 'html.parser')
```

```python
#Se revisa su contenido
print(f'page: \n{page.prettify()}')
```

```python
#PÁGINA FALABELLA
from bs4 import BeautifulSoup

#importar requests
import requests

#Abriendo página de falabella
url = 'https://www.falabella.com/falabella-cl/category/cat7090034/Tecnologia'

#Obtención de la página
page = requests.get(url)
print(f"Codigo: {page.status_code}")

page = BeautifulSoup(page.content, 'html.parser')
print(page.prettify())
```
